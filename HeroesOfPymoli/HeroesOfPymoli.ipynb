
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# calculation for purchase count, avg. purchase price, avg. total purchase per person etc. by gender
# extract columns by gender
genderPurchaseCount = purchase_data.groupby(["Gender"]).count()["Price"].rename("Purchase Count")

# avgerage amount per purchase per gender
genderAvgPurchase = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")

# total amount of purchases per gender
genderTotalPurchase = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")

# Calculate the average total purchase per person
avgTotalPurchasePerPerson = genderTotalPurchase / genderDivSum
```

```python
# Create the dataframe
genderDF = pd.DataFrame({
    "Purchase Count": genderPurchaseCount,
    "Average Purchase Price": genderAvgPurchase,
    "Total Purchase Value": genderTotalPurchase,
    "Average Total Purchase per Person": avgTotalPurchasePerPerson
})

# Formating the data for average purchase, total purchase value, and average total purchase per person
genderDF["Average Purchase Price"] = genderDF["Average Purchase Price"].map("${:,.2f}".format)
genderDF["Total Purchase Value"] = genderDF["Total Purchase Value"].map("${:,.2f}".format)
genderDF["Average Total Purchase per Person"] = genderDF["Average Total Purchase per Person"].map("${:,.2f}".format)

# Display the dataframe
genderDF
```

```python
# Retrieve the Item ID, Item Name, and Item Price columns
itemData = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
itemData
```

```python
# the purchase count
itemCounts = itemData.groupby(["Item ID", "Item Name"]).count()["Price"]
itemCounts
```

```python
# Calculate the average item price
avgItemPrice= itemData.groupby(["Item ID", "Item Name"]).mean()["Price"]

# Calculate total purchase value
totalItemPurchaseValue = itemData.groupby(["Item ID", "Item Name"]).sum()["Price"]
totalItemPurchaseValue
```
